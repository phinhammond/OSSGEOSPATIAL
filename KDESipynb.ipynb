```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", context="talk")


winter_file = "/content/drive/MyDrive/Winter_Biweekly_SAVI_Nov_to_Mar.csv"

winter = pd.read_csv(winter_file)
winter.head()

winter = winter.dropna(subset=["S2_SAVI"])
winter["S2_SAVI"] = pd.to_numeric(winter["S2_SAVI"], errors="coerce")
winter = winter.dropna(subset=["S2_SAVI"])

winter["AOI"].value_counts()

plt.figure(figsize=(10,6))

sns.kdeplot(
    data=winter,
    x="S2_SAVI",
    hue="AOI",
    fill=True,
    common_norm=False,
    alpha=0.35,
    linewidth=2,
    bw_adjust=.7
)

plt.title("Winter SAVI KDE  OSS Dump vs Split Creek")
plt.xlabel("SAVI")
plt.ylabel("Density")
plt.xlim(0, 1)
plt.show()



```
